### DATA ANALYSIS AND VISUALIZATION

```python
# Sample 100 rows from the DataFrame
food_sample = food.sample(n=100)
```

Participants by Nationality:

This bar chart shows the number of participants from each nationality. The x-axis represents the nationality, and the y-axis represents the number of participants. The height of each bar indicates the number of participants from each nationality. This visualization helps us understand the distribution of participants across different nationalities.

```python
# Bar chart for Nationality
plt.figure(figsize=(8,6))
sns.countplot(x='Nationality', data=food_sample)
plt.xticks(rotation=90)
plt.tight_layout()
plt.title('Participants by Nationality')
plt.show()
```

 Age Distribution:

This histogram shows the distribution of participants by age. The x-axis represents the age range, and the y-axis represents the number of participants. The height of each bar indicates the number of participants in each age range. This visualization helps us understand the age demographics of the participants.

```python
# Histogram for Age
plt.figure(figsize=(8,6))
sns.histplot(x='Age', data=food_sample, bins=10)
plt.title('Age Distribution')
plt.show()
```

Gender Distribution:

This bar chart shows the gender distribution of participants. The slices of the pie represent the proportion of male and female participants. This visualization helps us understand the gender balance of the participants.

```python
# bar chart for Gender
plt.figure(figsize=(6,6))
sns.countplot(x='Gender', data=food_sample)
plt.title('Gender Distribution')
plt.show()
```

Correlation between Food, Juice, and Dessert preferences:

This heatmap shows the correlation between the preferences for food, juice, and dessert. The x-axis and y-axis represent the different types of food, juice, and dessert, and the color of each cell indicates the strength of the correlation between each pair of preferences. This visualization helps us understand the relationships between the different preferences.

These visualizations provide insights into the demographics and preferences of the participants, which can help inform marketing strategies, product development, and other business decisions.

```python
# Stacked bar chart for Food, Juice, and Dessert preferences
plt.figure(figsize=(8,6))
sns.countplot(x='Food', data=food_sample, hue='Juice')
plt.title('Food and Juice Preferences')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Food', data=food_sample, hue='Dessert')
plt.title('Food and Dessert Preferences')
plt.show()
```

```python
# Group by Nationality and plot the mean Age
plt.figure(figsize=(8,6))
sns.barplot(x='Nationality', y='Age', data=food_sample)
plt.xticks(rotation=90)
plt.tight_layout()
plt.title('Mean Age by Nationality')
plt.show()
```

```python
# Group by Gender and plot the count of Food preferences
plt.figure(figsize=(8,6))
sns.countplot(x='Food', hue='Gender', data=food_sample)
plt.title('Food Preferences by Gender')
plt.show()
```

```python
# Group by Juice and plot the count of Dessert preferences
plt.figure(figsize=(8,6))
sns.countplot(x='Dessert', hue='Juice', data=food_sample)
plt.title('Dessert Preferences by Juice')
plt.show()
```

```python
# Group by Nationality and plot the count of Juice preferences
plt.figure(figsize=(8,6))
sns.countplot(x='Juice', hue='Nationality', data=food_sample)
plt.title('Juice Preferences by Nationality')
plt.show()
```

```python
# Group by Age and plot the count of Food preferences
plt.figure(figsize=(8,6))
sns.countplot(x='Food', hue='Age', data=food_sample)
plt.title('Food Preferences by Age')
plt.show()
```
